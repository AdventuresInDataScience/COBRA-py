# COBRA-py â€” Quickstart Example

This notebook walks through a complete COBRA-py run from raw data to a deployed, human-readable trading strategy.

**What we will do:**
1. Load and validate OHLCV data
2. Precompute all indicator arrays
3. Define a performance objective
4. Run the optimiser with a fixed seed (reproducible)
5. Inspect the best strategy in plain English
6. Run walk-forward validation and review out-of-sample performance
7. Run a second seed to see a diverse alternative strategy

---

> **MVP scope:** This example uses single-instrument OHLCV data only. External time-series data (e.g. policy rates) and portfolio-level optimisation are planned for later versions.

````python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# Core COBRA-py imports
from cobra_py.data.loader import load_ohlcv
from cobra_py.data.preprocessor import preprocess
from cobra_py.indicators.registry import DEFAULT_REGISTRY
from cobra_py.indicators.precompute import precompute_all
from cobra_py.search.space import build_config_space
from cobra_py.search.dehb_runner import run_dehb
from cobra_py.validation.walk_forward import walk_forward_validate
from cobra_py.reporting.report import generate_report, policy_to_human_readable

print('COBRA-py imports OK')
````

---

## 1. Load and Validate Data

`load_ohlcv` accepts a CSV or Parquet file path, or a pandas DataFrame directly. It validates all required columns, checks for data integrity, and ensures the index is a sorted DatetimeIndex.

For this example we generate a synthetic 10-year daily price series using geometric Brownian motion. In practice, replace this with your own data file.

````python
# --- Generate synthetic OHLCV data (replace with your own data) ---
np.random.seed(0)
n_bars = 2520  # ~10 years of daily bars
dates = pd.bdate_range(start='2014-01-01', periods=n_bars)

# Geometric Brownian Motion for close prices
log_returns = np.random.normal(loc=0.0003, scale=0.012, size=n_bars)
close = 100 * np.exp(np.cumsum(log_returns))

# Derive OHLV from close
daily_range = close * np.abs(np.random.normal(0, 0.008, n_bars))
high   = close + daily_range * np.random.uniform(0.3, 1.0, n_bars)
low    = close - daily_range * np.random.uniform(0.3, 1.0, n_bars)
open_  = close * (1 + np.random.normal(0, 0.003, n_bars))
volume = np.random.lognormal(mean=14, sigma=0.5, size=n_bars).astype(int)

raw_df = pd.DataFrame({
    'open':   open_,
    'high':   high,
    'low':    low,
    'close':  close,
    'volume': volume,
}, index=dates)

# --- Load and validate ---
data = load_ohlcv(raw_df)
print(f'Loaded {len(data):,} bars from {data.index[0].date()} to {data.index[-1].date()}')
data.tail(3)
````

---

## 2. Preprocess and Split

We split the data into an **in-sample** (training) set used for optimisation, and an **out-of-sample** (test) set held back for validation. The split is 70/30 by default.

````python
config = {
    'train_split': 0.70,
    'freq': '1D',
}

train, test = preprocess(data, config)
print(f'In-sample:  {len(train):,} bars  ({train.index[0].date()} â€“ {train.index[-1].date()})')
print(f'Out-of-sample: {len(test):,} bars  ({test.index[0].date()} â€“ {test.index[-1].date()})')
````

---

## 3. Precompute Indicators

All technical indicator arrays are computed **once** across all parameter combinations and stored in a cache. This is the most time-consuming step, but it only runs once. Every optimiser evaluation is then just an array lookup.

All indicators are computed using `pandas-ta` â€” no system-level library dependencies.

````python
print('Precomputing indicators... (this may take 1â€“3 minutes on first run)')

cache = precompute_all(
    data=train,
    registry=DEFAULT_REGISTRY,
    n_jobs=-1,  # use all available CPU cores
)

print(f'\nIndicator cache built.')
print(f'Memory usage: {cache.memory_usage_gb():.2f} GB')
print(f'Available indicators: {list(cache._store.keys())}')
````

---

## 4. Define the Search Space

The `build_config_space` function creates a **conditional** hyperparameter space. Each rule slot can be active or inactive, and the parameters available depend on which indicator is selected â€” for example, Bollinger Band rules expose `period`, `std`, `ma_type`, and `band_side`, while RSI rules expose only `period`.

The space is encoded using the `ConfigSpace` library and consumed directly by the DEHB optimiser.

````python
cs = build_config_space(
    n_entry_rules=3,   # up to 3 simultaneous entry conditions
    n_exit_rules=1,    # up to 1 rule-based exit condition
    registry=DEFAULT_REGISTRY,
)

print(f'Search space contains {len(cs.get_hyperparameters())} hyperparameters')
print(f'(Many are conditional â€” only a subset is active per configuration)')

# Quick sanity check: sample a few configurations and confirm they decode
from cobra_py.search.space import sample_and_validate
valid_configs = sample_and_validate(cs, n_samples=5)
print(f'\nSample validation: {len(valid_configs)} valid configs drawn')
````

---

## 5. Run the Optimiser

DEHB (Differential Evolution Hyperband) is used as the primary optimiser. It combines evolutionary search with multi-fidelity scheduling: poor strategies are evaluated on shorter data windows first and eliminated cheaply, so the compute budget is concentrated on promising candidates.

**The `seed` parameter controls reproducibility.** The same seed always produces the same result. Different seeds explore different regions of the search space.

Here we optimise for **Sharpe ratio**. Other options: `calmar`, `sortino`, `max_return`, `composite`.

````python
obj_config = {
    'objective': 'sharpe',    # what we are optimising for
    'complexity_penalty': 0.02,
    'min_trades': 10,
}

backtest_config = {
    'init_cash': 10000.0,
    'fee_rate': 0.001,     # 10bps per trade
    'slippage': 0.0005,
    'freq': '1D',
}

print(f"Running DEHB optimiser (budget=200 evaluations, seed=42)...")
print(f"Objective: {obj_config['objective']}")
print()

result = run_dehb(
    cache=cache,
    data=train,
    config_space=cs,
    obj_config=obj_config,
    backtest_config=backtest_config,
    budget=200,            # increase to 500+ for production runs
    min_fidelity=0.1,      # start evaluations on 10% of training data
    max_fidelity=1.0,      # final evaluations use 100% of training data
    n_workers=1,
    seed=42,               # â† reproducibility
)

print(f'Optimisation complete in {result.runtime_seconds:.1f}s')
print(f'Evaluations completed: {result.n_evaluations}')
print(f'Best in-sample Sharpe: {result.best_metrics["sharpe_ratio"]:.3f}')
````

---

## 6. Inspect the Best Strategy

The best strategy is expressed as a **human-readable rule set**. No model weights, no black box â€” just a set of named indicator conditions.

````python
print('=' * 60)
print('BEST STRATEGY FOUND')
print('=' * 60)
print()
print(policy_to_human_readable(result.best_policy))
print()
print('--- In-sample performance ---')
for k, v in result.best_metrics.items():
    if isinstance(v, float):
        print(f'  {k:<28} {v:.4f}')
    else:
        print(f'  {k:<28} {v}')
````

---

## 7. Out-of-Sample Evaluation

We now evaluate the best strategy on the held-out test set. **No refitting is performed** â€” this is a true out-of-sample test.

````python
from cobra_py.indicators.precompute import precompute_all as precompute
from cobra_py.backtest.engine import run_backtest

# Precompute indicators on the TEST set (using the SAME params as training)
test_cache = precompute(data=test, registry=DEFAULT_REGISTRY, n_jobs=-1)

# Evaluate the best policy on test data
oos_metrics = run_backtest(
    policy=result.best_policy,
    cache=test_cache,
    data=test,
    config=backtest_config,
)

print('--- Out-of-sample performance ---')
metrics_to_show = ['sharpe_ratio', 'calmar_ratio', 'total_return', 'max_drawdown', 'n_trades', 'win_rate']
for k in metrics_to_show:
    is_val = result.best_metrics.get(k, float('nan'))
    oos_val = oos_metrics.get(k, float('nan'))
    if isinstance(is_val, float):
        print(f'  {k:<28}  IS: {is_val:>8.3f}   OOS: {oos_val:>8.3f}')
    else:
        print(f'  {k:<28}  IS: {is_val:>8}   OOS: {oos_val:>8}')
````

---

## 8. Walk-Forward Validation

Walk-forward validation is the proper way to assess a rule-discovery system. It simulates the live deployment process: optimise on the past, trade on what comes next, roll forward, repeat.

This gives a realistic picture of how much the in-sample performance degrades out-of-sample across multiple time periods.

````python
def optimise_fn(train_data):
    """Wrapper that runs a full optimisation on a given training window."""
    fold_cache = precompute_all(train_data, DEFAULT_REGISTRY, n_jobs=-1)
    fold_cs = build_config_space(n_entry_rules=3, n_exit_rules=1, registry=DEFAULT_REGISTRY)
    return run_dehb(
        cache=fold_cache,
        data=train_data,
        config_space=fold_cs,
        obj_config=obj_config,
        backtest_config=backtest_config,
        budget=100,   # reduced for speed in this example
        seed=42,
    )

print('Running 3-fold walk-forward validation...')
print('(Each fold: optimise on past â†’ evaluate on next period)')
print()

wf_result = walk_forward_validate(
    data=data,
    optimise_fn=optimise_fn,
    config=config,
    n_splits=3,
    train_pct=0.70,
)

print('Walk-Forward Results:')
print(f'  OOS Sharpe â€” mean: {wf_result.oos_sharpe_mean:.3f}  std: {wf_result.oos_sharpe_std:.3f}')
print(f'  OOS Calmar â€” mean: {wf_result.oos_calmar_mean:.3f}')
print(f'  OOS Return â€” mean: {wf_result.oos_return_mean:.3f}')
print()
print('Per-fold out-of-sample Sharpe:')
for i, fold in enumerate(wf_result.fold_results):
    print(f'  Fold {i+1}: {fold["sharpe_ratio"]:.3f}')
````

## 9. Generate a Report

Save the full result as both a human-readable YAML file and a machine-readable JSON file.

````python
import os
os.makedirs('results', exist_ok=True)

generate_report(
    result=result,
    wf_result=wf_result,
    output_path='results/',
)

print('Report saved to results/')
print('  results/result.json       â€” machine-readable')
print('  results/result.yaml       â€” human-readable summary')
print('  results/best_policy.yaml  â€” strategy configuration')
````

---

## 10. Diversity: Run a Different Seed

Running with a different seed explores a different region of the search space and produces a qualitatively different strategy. This is by design â€” the two strategies can have different risk/return profiles and may be uncorrelated, making them suitable for combination into an ensemble.

The only change is `seed=123` instead of `seed=42`.

````python
print('Running second optimisation with seed=123...')
print()

result_2 = run_dehb(
    cache=cache,
    data=train,
    config_space=cs,
    obj_config=obj_config,
    backtest_config=backtest_config,
    budget=200,
    seed=123,              # â† different seed â†’ different strategy
)

print('Strategy from seed=42:')
print(policy_to_human_readable(result.best_policy))
print()
print('Strategy from seed=123:')
print(policy_to_human_readable(result_2.best_policy))
print()
print(f'Seed 42  â€” In-sample Sharpe: {result.best_metrics["sharpe_ratio"]:.3f}')
print(f'Seed 123 â€” In-sample Sharpe: {result_2.best_metrics["sharpe_ratio"]:.3f}')
````

---

## 11. Try a Different Objective

The objective is a first-class parameter. Changing it produces a strategy optimised for a different risk/return characteristic. Here we switch to `calmar` â€” maximising return per unit of max drawdown â€” which typically produces more conservative strategies with shallower drawdowns.

````python
calmar_config = {**obj_config, 'objective': 'calmar'}

print('Running optimisation with Calmar objective...')

result_calmar = run_dehb(
    cache=cache,
    data=train,
    config_space=cs,
    obj_config=calmar_config,
    backtest_config=backtest_config,
    budget=200,
    seed=42,
)

print()
print('Strategy optimised for Calmar ratio:')
print(policy_to_human_readable(result_calmar.best_policy))
print()

# Compare the two strategies
print('Comparison (in-sample):')
print(f'  {'Metric':<25}  {'Sharpe obj':>12}  {'Calmar obj':>12}')
print(f'  {"â”€" * 52}')
for k in ['sharpe_ratio', 'calmar_ratio', 'total_return', 'max_drawdown']:
    v1 = result.best_metrics.get(k, float('nan'))
    v2 = result_calmar.best_metrics.get(k, float('nan'))
    print(f'  {k:<25}  {v1:>12.4f}  {v2:>12.4f}')
````
